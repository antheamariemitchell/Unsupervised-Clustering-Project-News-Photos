```python
#Code for reading in dataframes and then doing PCA on images in each section

#Fox

#NYT

#BBC
BBCHealth = pd.read_csv('NYT_World_Images.csv')
```

```python
pca_reg = decomposition.PCA(6)
pca_reg.fit(BBCHealth.T)
PCA_BBC_Health = pca_reg.transform(BBCHealth.T)
```

```python
# Retrieve the first row of the dataframe
PCA_BBC_Health = pd.DataFrame(PCA_BBC_Health)
PCA_BBC_Health = PCA_BBC_Health.T
```
